```python
import pandas as pd
import folium

# Load the CSV files
visualize_df = pd.read_csv('visualize.csv')
nodes_visualize_df = pd.read_csv('nodes visualize.csv')

# Extract coordinates
visualize_coords = visualize_df[['Y', 'X']]
nodes_visualize_coords = nodes_visualize_df[['Y', 'X']]
pyeongtaek_logistics_center = [37.116718, 127.022459]

# Create a map centered around a central location
center_location = [(visualize_coords['Y'].mean() + nodes_visualize_coords['Y'].mean()) / 2,
                   (visualize_coords['X'].mean() + nodes_visualize_coords['X'].mean()) / 2]
map_ = folium.Map(location=center_location, zoom_start=10)

# Add visualize nodes in blue
for idx, row in visualize_coords.iterrows():
    folium.CircleMarker(location=(row['Y'], row['X']),
                        radius=5,
                        color='blue',
                        fill=True,
                        fill_color='blue',
                        fill_opacity=0.6).add_to(map_)

# Add nodes visualize nodes in red
for idx, row in nodes_visualize_coords.iterrows():
    folium.CircleMarker(location=(row['Y'], row['X']),
                        radius=5,
                        color='red',
                        fill=True,
                        fill_color='red',
                        fill_opacity=0.6).add_to(map_)

# Add Pyeongtaek Logistics Center in purple
folium.Marker(location=pyeongtaek_logistics_center,
              popup='Pyeongtaek Logistics Center',
              icon=folium.Icon(color='purple')).add_to(map_)

# Save the map to an HTML file
map_.save('node_visualization_map.html')

map_
```
